## Indicator 

```python
# depression
dp1= pd.read_sas('depression/dep09-10.XPT')
dp2 = pd.read_sas('depression/dep11-12.XPT')
dp3 = pd.read_sas('depression/dep13-14.XPT')
dp4 = pd.read_sas('depression/dep15-16.XPT')
dp5 = pd.read_sas('depression/dep17-18.XPT')
```

```python
dps = [dp1,dp2,dp3,dp4,dp5]

# concat
depression = pd.concat(dps, join='outer', axis=0).reset_index(drop=True)
```

```python
pd.set_option('display.float_format','{:.0f}'.format)
depression
```

```python
# Replace 7 or 9 to 0.
depression = depression.replace([7,9],[0, 0])
depression.DPQ010.unique()
```

```python
depression.isnull().sum()
```

```python
# Drop DPQ100, which has too many null values
depression=depression.drop(["DPQ100"], axis = 1)
```

```python
# Drop NAN
depression = depression.dropna()
```

```python
depression.isnull().sum()
```

```python
# sum all the columns except the first one.
depression['sum'] = depression.iloc[:, 1:].sum(axis=1)
depression
```

```python
# add an indicator column, if the sum greater than 8, print 1. Else print 0.
depression['indicator'] = [1 if x >=8 else 0 for x in depression['sum']]
```

```python
# Keep only the SEQN and indicator column
depression_indicator = depression[['SEQN', 'indicator']]
```

```python
# depression_indicator.to_csv('depression_indicator.csv', index=False)
```

```python
# Concat all the features.
from functools import partial, reduce

# Concat all questionnaire data
dfs = [weight,health,activity]
merge = partial(pd.merge, on='SEQN', how='outer')
features = reduce(merge, dfs)
```

```python
# Concat demographic data and questionnaire data
dfs2 = [features,demographic]
merge = partial(pd.merge, on='SEQN', how='outer')
final = reduce(merge, dfs2)
final
```

```python
# left merge final datafram with indicator
final_data = pd.merge(depression_indicator,final,how='left', on='SEQN')
final_data
```

fill nulls 

For citizenship, and Education, fill Mode.

For Household Income, fill Median.

For marital_status, fill nan with 0 indicates missing.

```python
final_data['Citizenship'].fillna(int(final_data['Citizenship'].mode()), inplace=True)
final_data['Education'].fillna(int(final_data['Education'].mode()), inplace=True)

final_data['Household_Income'].fillna(int(final_data['Household_Income'].median()), inplace=True)

final_data['Marital_Status'] = final_data['Marital_Status'].fillna(77)
```
